```python
import matplotlib.pyplot as plt
import seaborn as sns
iris=sns.load_dataset("iris")
```

```python
print(iris)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plot = sns.histplot(data=iris, x="sepal_width", bins=10, kde=True, color="skyblue")
plot.set_title("Histogram")
plot.set_xlabel("Sepal Width")
plot.set_ylabel("Frequency")


plt.subplot(1,2,2)
plot = sns.boxplot(data=iris, x="sepal_width", color="lightcoral")
plot.set_title("Boxplot")
plot.set_xlabel("Sepal Width")

plt.tight_layout()
plt.show()

```
